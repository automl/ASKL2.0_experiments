```python
tasks = openml.tasks.list_tasks(task_type_id=1, output_format='dataframe')
tasks = tasks.loc[task_ids]
tasks
```

```python
task_ids_sorted_by_num_features = sorted(task_ids, key=lambda tid: tasks.loc[tid, 'NumberOfInstances'])
```

```python
for tid in task_ids_sorted_by_num_features:
    plt.figure(figsize=[16,12])
    colors = itertools.cycle(['blue', 'orange', 'green', 'red', 'purple', 'grey', 'cyan',
                              'magenta', 'black', 'brown'])
    markers = itertools.cycle(['o', 'v', '^', '<', '>', '8', '*', '+', 'd'])
    for mode in list(valid_pretty.keys()):
        c = next(colors)
        m = next(markers)
        tmp = pd.DataFrame(res_dc[tid][mode]).sort_index(axis=1).ffill(axis=1)
        med = tmp.median(axis=0)
        med.loc[1200] = med.iloc[-1]
        low = tmp.quantile(0.25)
        low.loc[1200] = low.iloc[-1]
        up = tmp.quantile(0.75, axis=0)
        up.loc[1200] = up.iloc[-1]
        plt.plot(med.index, med.to_numpy(), label=valid_pretty[mode], linewidth=3, 
                 #color=c, marker=m
                )
        plt.fill_between(med.index, low, up, alpha=0.3, 
                         #color=c
                        )
    plt.title('Name: %s (%d), #instances: %d, #attributes: %d' % (
        tasks.loc[tid, 'name'], tid, tasks.loc[tid, 'NumberOfInstances'], tasks.loc[tid, 'NumberOfFeatures']))
    plt.legend(fontsize=20)
    plt.ylim([plt.ylim()[0], plt.ylim()[0] + 0.3*(plt.ylim()[1]-plt.ylim()[0])])
    #plt.yscale("log")
    plt.show()
```

```python
plt.figure(figsize=[16,12])
bounds = [0, 30]

for lower_bound in bounds:
    all_results = collections.defaultdict(list)
    alpha = (lower_bound + 10) / (bounds[-1] + 10)
    for tid in task_ids_sorted_by_num_features[lower_bound:]:
        if tid in [189873, 189866]:
            continue
        min_for_task = 1.0
        max_for_task = 0.0
        for mode in list(valid_pretty.keys()):
            tmp = pd.DataFrame(res_dc[tid][mode]).sort_index(axis=1).ffill(axis=1)
            mini = tmp.min().min()
            min_for_task = min(min_for_task, mini)
            maxi = tmp.max().max()
            max_for_task = max(max_for_task, maxi)
        diff = max_for_task - min_for_task
        if diff == 0.0:
            diff = 1.0
        for mode in list(valid_pretty.keys()):
            tmp = pd.DataFrame(res_dc[tid][mode]).sort_index(axis=1).ffill(axis=1).mean()
            tmp = (tmp - min_for_task) / diff
            all_results[mode].append(tmp)

    colors = itertools.cycle(['blue', 'orange', 'green', 'red', 'purple', 'grey', 'cyan',
                              'magenta', 'black', 'brown'])
            
    for mode in list(valid_pretty.keys()):
        c = next(colors)
        tmp = pd.DataFrame(all_results[mode]).ffill(axis=1)
        tmp = tmp.mean()
        if lower_bound == bounds[-1]:
            label=valid_pretty[mode]
        else:
            label = None
        plt.plot(tmp, label=label, linewidth=3, alpha=alpha, c=c)
    
plt.legend(fontsize=20)
plt.ylim((0.01, 0.2))
#plt.yscale('log')
```

```python
plt.figure(figsize=[16,12])

all_results = collections.defaultdict(list)
for tid in task_ids_sorted_by_num_features:
    if tid in [189873, 189866]:
        continue
    min_for_task = 1.0
    max_for_task = 0.0
    for mode in list(valid_pretty.keys()):
        tmp = pd.DataFrame(res_dc[tid][mode]).sort_index(axis=1).ffill(axis=1)
        mini = tmp.min().min()
        min_for_task = min(min_for_task, mini)
        maxi = tmp.max().max()
        max_for_task = max(max_for_task, maxi)
    diff = max_for_task - min_for_task
    if diff == 0.0:
        diff = 1.0
    for mode in list(valid_pretty.keys()):
        tmp = pd.DataFrame(res_dc[tid][mode]).sort_index(axis=1).ffill(axis=1).mean()
        tmp = (tmp - min_for_task) / diff
        all_results[mode].append(tmp.iloc[-1])

colors = itertools.cycle(['blue', 'orange', 'green', 'red', 'purple', 'grey', 'cyan',
                          'magenta', 'black', 'brown'])

to_plot = np.ones((len(task_ids_sorted_by_num_features), len(valid_pretty))) * np.NaN
for i, tid in enumerate(task_ids_sorted_by_num_features):
    if tid in [189873, 189866]:
        continue
    for j, mode in enumerate(list(valid_pretty.keys())):
        min_idx = max(0, i - 2)
        max_idx = min(i + 3, len(task_ids_sorted_by_num_features))
        values = all_results[mode][min_idx: max_idx]
        to_plot[i, j] = np.nanmean(values)
        
#for mode in list(valid_pretty.keys()):
#    print(mode, all_results[mode])
#print(to_plot)

for i, mode in enumerate(list(valid_pretty.keys())):
    c = next(colors)
    tmp = to_plot[:, i].flatten()
    if lower_bound == bounds[-1]:
        label=valid_pretty[mode]
    else:
        label = None
    plt.plot(tmp, label=label, linewidth=3, c=c)
    
plt.legend(fontsize=20)
plt.ylim((0.0, 0.1))
```
